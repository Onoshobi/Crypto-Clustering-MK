```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
df_market_data.columns
```

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
from sklearn.preprocessing import StandardScaler

features = ['price_change_percentage_24h', 'price_change_percentage_7d',
       'price_change_percentage_14d', 'price_change_percentage_30d',
       'price_change_percentage_60d', 'price_change_percentage_200d',
       'price_change_percentage_1y']

sc = StandardScaler()
sc_feature_arrays = sc.fit_transform(df_market_data[features])
```

```python
df_market_data.index
```

```python
# Create a DataFrame with the scaled data
# Copy the crypto names from the original DataFrame

df_market_data_scaled = pd.DataFrame(sc_feature_arrays,columns=features)

df_market_data_scaled.reset_index()
# Set the coin_id column as index
df_market_data_scaled.set_index(df_market_data.index,inplace=True)

# Display the scaled DataFrame
df_market_data_scaled.head(50)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
no_cluster_list = [i for i in range(1,12)]
no_cluster_list
```

```python
# Create an empty list to store the inertia values
inertia_values_list = []

# Create a for loop to compute the inertia with each possible value of k

for test_k in no_cluster_list:
    # Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    k_means_model = KMeans(n_clusters=test_k,init="k-means++",random_state=0)
# 2. Fit the model to the data using `df_market_data_scaled`
    k_means_model.fit(df_market_data_scaled)
# 3. Append the model.inertia_ to the inertia list
    inertia_values_list.append(k_means_model.inertia_)


```

```python
inertia_values_list
```

```python
# Create a dictionary with the data to plot the Elbow curve
number_cluster_inertia_value_dict = { 'number_of_cluster':no_cluster_list,
                                     'inertia_based_on_no_cluster':inertia_values_list
                                    }

# Create a DataFrame with the data to plot the Elbow curve
df_cluster_data = pd.DataFrame(number_cluster_inertia_value_dict)
df_cluster_data
```

```python
# Plot a line chart with all the inertia values computed with

# the different values of k to visually identify the optimal value for k.


elbow_curve_origional_df = df_cluster_data.hvplot.line(
    x='number_of_cluster',
    y='inertia_based_on_no_cluster',
    title="Elbow Curve For Original Data",
    width=400,
    height=300
)
elbow_curve_origional_df
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** We have to select the value of k at the “elbow” that is the point after which the distortion/inertia starts decreasing in a linear fashion. Thus for the given data, we conclude that the optimal number of clusters for the data is 4.

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

---

```python
# Initialise the K-Means model using the best value for k
k_means_model = KMeans(n_clusters=4,init="k-means++",random_state=0)
```

```python
# Fit the K-Means model using the scaled DataFrame
k_means_model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
array_cluster_value = k_means_model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
print(array_cluster_value)
```

```python
# Create a copy of the scaled DataFrame
copy_df_market_data_scaled = df_market_data_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
copy_df_market_data_scaled['predicted_cluster'] = array_cluster_value

# Display the copy of the scaled DataFrame
copy_df_market_data_scaled.head(50)
```

```python
copy_df_market_data_scaled.reset_index(inplace=True)
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Colour the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
x="price_change_percentage_24h"
y="price_change_percentage_7d"
scatter_plot_orginal_scaled_df = copy_df_market_data_scaled.hvplot.scatter(x=x, y=y,
                  legend='top',height=300,width=400,by='predicted_cluster', 
                  hover_cols=['coin_id','predicted_cluster'],title="Scatter Plot Original Data")

scatter_plot_orginal_scaled_df
```

### Optimise Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
```

```python
df_market_data_scaled.columns
```

```python
features
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled Dataframe
# down to three principal components.
pca_3_component = pca.fit_transform(df_market_data_scaled)

# View the scaled PCA data
pca_3_component
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the three principal components is 0.895.

```python
# Create a new DataFrame with the PCA data.
column_name = ["PC1","PC2","PC3"]
df_market_data_pca = pd.DataFrame(pca_3_component,columns=column_name)



# Copy the crypto names from the original scaled DataFrame
crypto_name = df_market_data_scaled.index
# Set the coin_id column as index
df_market_data_pca.set_index(crypto_name,inplace=True)


# Display the scaled PCA DataFrame
df_market_data_pca.head(10)
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
number_of_cluster_list = [i for i in range(1,12)]
number_of_cluster_list
```

```python
# Create an empty list to store the inertia values
inertia_values_list_pca = []

# Create a for loop to compute the inertia with each possible value of k

# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list

for clusters in number_of_cluster_list:
    # Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    k_means_model = KMeans(n_clusters=clusters,init="k-means++",random_state=0)
# 2. Fit the model to the data using `df_market_data_scaled`
    k_means_model.fit(df_market_data_pca)
# 3. Append the model.inertia_ to the inertia list
    inertia_values_list_pca.append(k_means_model.inertia_)
inertia_values_list_pca
```

```python
# Create a dictionary with the data to plot the Elbow curve
number_cluster_inertia_value_dict_pca = { 'number_of_cluster':number_of_cluster_list,
                                     'inertia_based_on_no_cluster':inertia_values_list_pca
                                    }

# Create a DataFrame with the data to plot the Elbow curve
df_cluster_data_pca = pd.DataFrame(number_cluster_inertia_value_dict_pca)
df_cluster_data_pca
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_curve_pca_df = df_cluster_data_pca.hvplot.line(
    x='number_of_cluster',
    y='inertia_based_on_no_cluster',
    title="Elbow Curve After PCA",
    width=400,
    height=300
)
elbow_curve_pca_df

```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** we have to select the value of k at the “elbow” that the point after which the distortion/inertia starts decreasing in a linear fashion. Thus for the given data, we conclude that the optimal number of clusters for the data is 4.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** It is the same as previous that is k value 4 because “elbow” is the point after which the distortion/inertia starts decreasing in a linear fashion after 4 only.
